```python
# read in one recipe:




endpoint = 'http://localhost:8000'

q = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
PREFIX schema: <http://schema.org/>
PREFIX dapro: <http://dapro.opendata.zhaw.ch/DaPro.owl#>

SELECT ?ingredientmat ?ingredientmatval ?ingredientmatunit ?poportioninproduct
WHERE {

    ?dapro rdfs:label "Bio · Tofu · Curry".
   	?dapro schema:ingredients ?recingredient .

    ?recingredient schema:ingredients ?foodoningredient .
    ?recingredient schema:value ?foodoningredientval .
	?CHFDMingredient rdfs:isDefinedBy ?foodoningredient .
    ?CHFDMingredient rdfs:isDefinedBy ?otheringredients .
    ?otheringredients  <http://schema.org/ingredients> ?material.
    ?material schema:ingredients ?ingredientmat . 
    
    OPTIONAL{

    ?material schema:value ?ingredientmatval .
    ?material schema:unitCode ?ingredientmatunit .

    }
    BIND(?ingredientmatval * ?foodoningredientval AS ?poportioninproduct)


}
"""


df = sparql_dataframe.get(endpoint, q)
df
```

```python
df_nutrients = df.groupby('ingredientmat')['poportioninproduct', 'ingredientmatunit'].agg({'poportioninproduct':'sum', 'ingredientmatunit' : 'last'})
```
